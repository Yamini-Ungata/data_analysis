```python
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)
#array of zeros in integer type
a=np.zeros((3,3),dtype=int)
print(a)
#array of ones in integer type
b=np.ones((2,2),dtype=int)
print(b)

c=np.arange(10)
print(c)

#array manipulation
d=arr.reshape(5,1)
print(d)

#slicing the array
e=arr[2:4]
print(e)

#stacking of an array
#vertical stacking
a1=np.array([2,3,4,5,6])
a2=np.array([7,8,9,10,11])
r1=np.vstack(a1+a2)
print(r1)

#horizontal stacking
a1=np.array([2,3,4,5,6])
a2=np.array([7,8,9,10,11])
r1=np.stack(a1+a2)
print(r1)

a=np.array([3,4,5,6,7,8])
b=np.split(a,3)
print(b)

#tranpose of a matrix
a=np.array([[1,2,3],[4,5,6]])
b=a.T
print(b)
```

```python
#element wise operation
import numpy as np
a=np.array([3,5,6,7,9,1])
b=np.array([4,3,1,7,9,8])
c=a+b
print(c)
```

```python
#Broadcast operation
import numpy as np
a=np.array([1,2,3,4,5])
b=a+3
print(b)
```

```python
#Linear algebra with Numpy

#matrix multiplication
a=np.array([[2,3],[4,6]])
b=np.array([[2,1],[8,7]])
c=np.dot(a,b)
print(c)

#calculation of eigen values and eigen vectors
a=np.array([[2,3],[4,6]])
b=np.array([[2,1],[8,7]])
c=np.dot(a,b)
d=np.linalg.eig(c)
print(d)

#sum of elements in matrices
a=np.array([[2,3,4],[4,5,6]])
b=np.sum(a)
print(b)
d=np.array([[1,2,3],[4,8,7]])
c=np.sum(a+d)
print(c)


a=np.array([[2,3,4],[4,5,6]])
b=np.sum(a,axis=0)#rowwise operation
c=np.sum(a,axis=1)#columnwise operation
print(b)
print(c)


```

```python
#statistical operations
import numpy as np
a=np.array([6,5,4,3,2])
b=np.mean(a)
c=np.median(a)
v=np.var(a)
d=np.std(a)
print('mean',b)
print('median',c)
print('variance',v)
print('standard deviation',d)
```

```python
import matplotlib.pyplot as plt
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]

profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='CompanyA',marker='+',ms='15',mec='g')
plt.legend(loc='best')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='violet',linestyle='dotted',linewidth='3',label='CompanyB',marker='*',ms='15',mec='b')
plt.legend(loc='best')
```
