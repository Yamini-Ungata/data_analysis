**Extracting HTML code of any website using BeautifulSoup.**


```python
from bs4 import BeautifulSoup
import requests
```

```python
var1='https://www.wikipedia.org/'
```

```python
page=requests.get(var1)
```

```python
var2=BeautifulSoup(page.text,'html')
print(var2)
```

```python
t_var=var2.find('title')
print(t_var)
```

```python
print(t_var.text)
```

```python
a='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
page_var=requests.get(a)
```

```python
code_var=BeautifulSoup(page_var.text,'html')
print(code_var)
```

```python
table_var=code_var.find('table')
print(table_var.text)
```

```python
headings_var=table_var.find_all('th')
print(headings_var)
```

```python
#List comprehension Syntax
#List_name=[output for_loop]
```

```python
head=[i.text for i in headings_var]
print(head)
```

```python
import pandas as pd
```

```python
#Creating data frame
#DataFrame_name=pd.DataFrame(columns='data')
```

```python
df_var=pd.DataFrame(columns=head)
print(df_var)
```

```python
data_var=table_var.find_all('tr')
print(data_var)
```

```python
rows_var=[data_var[i].text for i in range(1,len(data_var)-1)]
print(rows_var)
```

```python
ind=0
for i in data_var[1:6]:
  #print(i)
  erd=i.find_all('td')
  #print(erd)
  row_data=[i.text for i in erd]
  #print(row_data)
  df_var.loc[ind]=row_data
  ind+=1
```
